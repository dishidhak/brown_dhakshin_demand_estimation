```python
df = pd.read_csv("air_fryers_clean_brand_year.csv")
```

```python
import statsmodels.api as sm
import pandas as pd

feature_cols = ['compact_share', 'dual_basket_share', 'oven_style_share',
                'rotisserie_share', 'window_share']

df_reg = pd.get_dummies(df, columns=['brand', 'year'], drop_first=True)

X_cols = ['avg_price', 'avg_rating'] + feature_cols + \
         [c for c in df_reg.columns if c.startswith('brand_') or c.startswith('year_')]

X = sm.add_constant(df_reg[X_cols].astype(float))
y = df_reg['log_brand_share'].astype(float)

model = sm.OLS(y, X).fit()
```

```python
print("Estimated price coefficient:", model.params['avg_price'])
```

The estimated price coefficient is negative, meaning higher prices are associated with lower market share. This is important because it is consistent with the basic economic principle that as prices increase, demand for that product goes down. If the coefficient were positive, it would imply that higher prices increase demand, which is unrealistic for air fryers and would make the model unreliable.

```python
model.params[feature_cols].sort_values(ascending=False)
```

Dual-basket, compact, and rotisserie features are associated with higher demand, as they have positive coefficients. In contrast, oven-style and window features have negative coefficients, indicating lower demand. This suggests that consumers tend to prefer more compact and multifunctional designs, while certain specialized features may be less desirable in this market.

```python
model.params.filter(like='brand_').sort_values(ascending=False)
```

The brands with the largest positive dummy coefficients are Cuisinart and Oster, followed by NuWave and Ninja. These coefficients indicate that, relative to the omitted baseline brand, these brands have higher market share after controlling for price, ratings, and product features.

In contrast, brands such as Cosori, GoWISE USA, and Dash have negative coefficients, indicating lower market share relative to the baseline brand.

```python
model.params.filter(like='year_').sort_values(ascending=False)
```

The year with the largest dummy coefficient is 2021, followed by 2023 and 2022. These positive coefficients indicate that, relative to the omitted baseline year (2019), overall demand was higher in these years. This suggests that the air fryer market experienced increased demand after 2019, with 2021 showing the strongest increase.

```python
print("R-squared:", model.rsquared)
```

About 89.1% of the variation in log market share is explained by the model. This suggests a strong fit.
